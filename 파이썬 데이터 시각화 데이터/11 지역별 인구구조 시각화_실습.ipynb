```python
fig, ax1 = plt.subplots()

#ÎßâÎåÄÍ∑∏ÎûòÌîÑ
ax1.bar(df_sido['ÏãúÎèÑÎ™Ö'],df_sido['Ï¥ùÏù∏Íµ¨Ïàò'],width=-0.4
       , align='edge', label='Ï¥ùÏù∏Íµ¨')
ax1.bar(df_sido['ÏãúÎèÑÎ™Ö'],df_sido['ÏÑ∏ÎåÄÏàò'],width=0.4
       , align='edge', label='ÏÑ∏ÎåÄÏàò')
ax1.legend(loc=(0.3, 0.8))
ax1.set_xlabel('ÏãúÎèÑÎ™Ö')
ax1.set_ylabel('Ï¥ùÏù∏Íµ¨Ïàò, ÏÑ∏ÎåÄÏàò')
plt.xticks(rotation=45)

#Í∫æÏùÄÏÑ†Í∑∏ÎûòÌîÑ
ax2 = ax1.twinx()

ax2.plot(df_sido['ÏãúÎèÑÎ™Ö'],df_sido['ÏÑ∏ÎåÄÎãπ Ïù∏Íµ¨'], 'k^-.',
        mfc='r', mec='r', label='ÏÑ∏ÎåÄÎãπ Ïù∏Íµ¨')
ax2.legend(loc=(0.6, 0.85))
ax2.set_ylabel('ÏÑ∏ÎåÄÎãπ Ïù∏Íµ¨')
ax2.set_title('2021ÎÖÑ 10Ïõî ÏãúÎèÑÎ≥Ñ Ï¥ù Ïù∏Íµ¨Ïàò, ÏÑ∏ÎåÄÏàò, ÏÑ∏ÎåÄÎãπ Ïù∏Íµ¨', size=20)

plt.show()
```
