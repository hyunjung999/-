# Ïó∞ÎèÑÎ≥Ñ ÏäπÍ∞ùÏàòÏùò Î≥ÄÌôî

## matplotlibÏúºÎ°ú ÏãúÍ∞ÅÌôî
* plt.plot(x,y)

### Îç∞Ïù¥ÌÑ∞ Í∞ÄÍ≥µ

```python
flights_year = flights.groupby('year')['passengers'].sum()
flights_year
```

### ÏãúÍ∞ÅÌôî

```python
plt.plot(flights_year, 'ro--')
plt.xlabel('year')
plt.ylabel('passengers')
plt.show()
```

## seaborn
* `sns.lineplot(data=Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ, x=xÏ∂ïÏª¨Îüº, y=yÏ∂ïÏª¨Îüº, estimator=ÌÜµÍ≥ÑÌï®Ïàò)`<br>
estimatorÎ•º ÏÉùÎûµÌïòÎ©¥ ÌèâÍ∑†ÏúºÎ°ú ÌÜµÍ≥ÑÎ•º Ï†ÅÏö©Ìï®

* Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞Î°ú Ï∞®Ìä∏Î•º Í∑∏Î¶¨Î©¥ Ïã†Î¢∞Íµ¨Í∞Ñ ÌëúÏãú

```python
sns.lineplot(data=flights, x='year', y='passengers', ci=None, estimator=sum, color='r', marker='p', ls=':')
plt.show()
```

# Ïó∞ÎèÑ-ÏõîÎ≥Ñ ÏäπÍ∞ùÏàòÏùò Î≥ÄÌôî

## matplotlibÏúºÎ°ú ÏãúÍ∞ÅÌôî

### Îç∞Ïù¥ÌÑ∞ Í∞ÄÍ≥µ

```python
flights_pivot = flights.pivot(index='year', columns='month', values='passengers')
flights_pivot
```

üí°Í∞Å Ïó∞ÎèÑÎ≥Ñ Îç∞Ïù¥ÌÑ∞Î•º ÌñâÏúºÎ°ú Î∂àÎü¨Ïò¨ Ïàò ÏûàÍ≤å ÎêòÏóàÎã§.

### ÏãúÍ∞ÅÌôî

```python
for i in range(12):
    plt.plot(flights_pivot.iloc[i], label=flights_pivot.index[i])

plt.legend()
plt.show()
```

üí°Îß§ÎÖÑ 7-8ÏõîÏóê ÏäπÍ∞ùÏàòÍ∞Ä Í∞ÄÏû• ÎßéÎã§.
