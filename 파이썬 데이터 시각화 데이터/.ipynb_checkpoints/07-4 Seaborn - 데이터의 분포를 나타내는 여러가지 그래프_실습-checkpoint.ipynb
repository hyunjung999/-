```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

# Ïπ¥Ïö¥Ìä∏ÌîåÎ°Ø
* Îç∞Ïù¥ÌÑ∞Î•º Í∑∏Î£πÌïëÌïòÏó¨ Í∞ØÏàòÎ•º Ïπ¥Ïö¥Ìä∏ÌïòÏó¨ ÏãúÍ∞ÅÌôî 
* `sns.countplot(data=Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ, x=Ïª¨Îüº)`<br>


```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Ïó∞Î†πÎ≥Ñ ÏäπÏÑ†Ïù∏Ïõê Ïπ¥Ïö¥Ìä∏ÌïòÏó¨ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(15, 4))
sns.countplot(data=titanic, x='age')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

üí°24ÏÑ∏Í∞Ä Í∞ÄÏû• ÎßéÏù¥ ÌÉëÏäπÌïú Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§. <br>
üí°ÏÜåÏàòÏ†êÏúºÎ°ú ÎêòÏñ¥ÏûàÎäî Ïù¥ÏÉÅÌïú ÎÇòÏù¥Í∞Ä Ï°¥Ïû¨ÌïúÎã§.

```python
# 1. ÎÇ®Ïûê/Ïó¨Ïûê/Ïñ¥Î¶∞Ïù¥ ÏäπÏÑ†Ïù∏Ïõê ÏãúÍ∞ÅÌôî
#sns.countplot(data=titanic, x='who')

# 2. ÎÇ®Ïûê/Ïó¨Ïûê/Ïñ¥Î¶∞Ïù¥ Î≥Ñ ÏÉùÏ°¥Ïó¨Î∂Ä 
sns.countplot(data=titanic, x='who', hue='alive')
plt.show()
```

üí°ÎÇ®ÏÑ± ÌÉëÏäπÍ∞ùÏùò ÏàòÍ∞Ä ÎßéÎã§. <br>
üí°ÎÇ®ÏÑ± ÌÉëÏäπÍ∞ùÏùò ÎåÄÎ∂ÄÎ∂ÑÏùÄ ÏÉùÏ°¥ÌïòÏßÄ Î™ªÌñàÎã§.

# Îü¨Í∑∏ÌîåÎ°Ø
* `sns.rugplot(data=Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ, x=Ïª¨Îüº)` <br>
Îç∞Ïù¥ÌÑ∞Ïùò ÏúÑÏπòÎ•º Î≥¥Ïó¨Ï§ÄÎã§.

```python
sns.rugplot(data=titanic, x='age', hue='alive')
plt.show()
```

# ÌûàÏä§ÌÜ†Í∑∏Îû®
* `sns.distplot(data=Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ, x=Ïª¨Îüº)` <br>
Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨Î•º ÏïåÎ†§Ï§ÄÎã§.

```python
iris = sns.load_dataset('iris')
iris.head(1)
```

```python
sns.displot(data=iris, x='petal_length', bins=20, rug=True, hue='species', kde=True)
# rug: Îü¨Í∑∏ÌîåÎ°Ø, kde: Î∞ÄÎèÑ Ìï®Ïàò Í≥°ÏÑ† Ï∂îÍ∞Ä 

plt.show()
```

```python
sns.displot(data=iris, x='petal_length', rug=True, hue='species', kind='kde')
# kindÎ°ú Ï¢ÖÎ•ò ÌëúÏãú
plt.show()
```

üí°ÎßâÎåÄÍ∑∏ÎûòÌîÑÏóÜÏù¥ Í≥°ÏÑ†ÏúºÎ°úÎßå Î∂ÑÌè¨Î•º ÌôïÏù∏Ìï† Ïàò ÏûàÎã§.

# ÏÉÅÏûêÏàòÏóºÍ∑∏ÎûòÌîÑ, Î∞îÏù¥Ïò¨Î¶∞ÌîåÎ°Ø, Ïä§Ìä∏Î¶ΩÌîåÎ°Ø, Ïä§ÏõúÌîåÎ°Ø
* sns.boxplot(data=Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ)
* sns.violinplot(data=Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ)
* sns.stripplot(data=Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ)
* sns.swarmplot(data=Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ)

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
sns.boxplot(data=iris)

plt.subplot(222)
sns.violinplot(data=iris)

plt.subplot(223)
sns.stripplot(data=iris)

plt.subplot(224)
sns.swarmplot(data=iris, s=2)

plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
# Î∞ïÏä§ÌîåÎ°Ø
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.subplot(222)
# Î∞îÏù¥Ïò¨Î¶∞ÌîåÎ°Ø
sns.violinplot(data=iris, x='species', y='sepal_length')
plt.subplot(223)
# Ïä§Ìä∏Î¶ΩÌîåÎ°Ø
sns.stripplot(data=iris, x='species', y='sepal_length')
plt.subplot(224)
# Ïä§ÏõúÌîåÎ°Ø
sns.swarmplot(data=iris, x='species', y='sepal_length', s=2)
plt.show()
```
